```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import jaccard_similarity_score
from scipy import stats
import pandas as pd
import numpy as np
import kmeans as KM
import math
%matplotlib inline
```

```python
def loadDataSet(fileName):
    dataSet=[]
    label_true=[]
    fr=open(fileName)
    ctoi={
        'Iris-setosa':0,
        'Iris-versicolor':1,
        'Iris-virginica':2
    }
    for line in fr.readlines()[:-1]:
        data=line.strip().split(',')
        dataSet.append(data[:-1])
        label_true.append(ctoi[data[-1]])
        dataSet=[map(float,data) for data in dataSet]
    return dataSet,label_true
```

```python
def testKM(dataArr,dataMat,k,label_true):
    rand_index=[]
    jaccard=[]
    label_pred=[]
    for t in range(50):
        kMeans = KMeans(n_clusters=k,init=np.array(KM.randCent(dataMat,k)),n_init=1).fit(dataArr)
        label_pred=list(kMeans.labels_)
        a=0
        b=0
        l=len(label_pred)
        sum=float(l*(l-1)/2)
        for i in range(l):
            for j in range(i,l):
                if label_true[i]==label_true[j] and label_pred[i]==label_pred[j]: a+=1
                if label_true[i] != label_true[j] and label_pred[i] !=label_pred[j] :b+=1
        rand_index.append(float(a+b)/sum)
        jaccard.append(float(a)/(sum-b))
    return rand_index,jaccard,label_pred
```

```python
def testBKM(datamat,k,label_true):
    rand_index=[]
    jaccard=[]
    t=0
    label_pred=[]
    while(t<50):
        a=0
        b=0
        l=len(label_pred)
        try:
            centroids, clusterAssment=KM.biKmeans(datamat,k)
            label_pred=list(clusterAssment.A[:,0].astype(pd.np.int))
            flag=0
            for center in centroids:
                if math.isnan(np.array(center)[0][0]): flag=1
            if flag:continue
            sum=float(l*(l-1)/2)
            for i in range(l):
                for j in range(i,l):
                    if (label_true[i]==label_true[j]) and (label_pred[i]==label_pred[j]): a+=1
                    if (label_true[i] != label_true[j]) and (label_pred[i] !=label_pred[j]) :b+=1
            rand_index.append(float(a+b)/sum)
            jaccard.append(float(a)/(sum-b))
        except:
            print "something wrong happen"
            continue
        t+=1
    return rand_index,jaccard,label_pred
```

```python
def statistics(dataSet,label_true):
    dataMat=np.mat(dataSet)
    f = pd.DataFrame(dataMat, columns=['sepal length', ' sepal width', ' petal length','petal width'])
    f['label']=label_true
    f_setosa=f[f.label==0]
    f_versicolor=f[f.label==1]
    f_virginica=f[f.label==2]
    print 'the count of setosa:',len(f_setosa)
    print 'the count of versicolor',len(f_versicolor)
    print 'the count of virginica',len(f_virginica)
    print 'means'
    print 'Iris\n',f.mean()
    print 'setosa\n',f_setosa.mean()
    print 'versicolor\n',f_versicolor.mean()
    print 'virginica\n',f_virginica.mean()
    print 'var'
    print 'Iris\n',f.var()
    print 'setosa\n',f_setosa.var()
    print 'versicolor\n',f_versicolor.var()
    print 'virginica\n',f_virginica.var()
```

```python
dataSet,label_true=loadDataSet('./iris.data')
dataArr=np.array(dataSet)
dataMat=np.mat(dataSet)
statistics(dataSet,label_true)
```

```python
f = pd.DataFrame(dataMat, columns=['sepal length', ' sepal width', ' petal length','petal width'])
f['label'] = label_true
color = ['red','blue', 'yellow', 'green', 'orange']
pd.scatter_matrix(f.iloc[:,:-1], figsize=(9, 9), diagonal='kde', marker='o', s=40, alpha=.4, c=f['label'].apply(lambda x: color[x]))
```

```python
res_kMeans=[]
res_biKMeans=[]
for i in range(2,5):
    res_kMeans.append(testKM(dataArr,dataMat,i,label_true))
for i in range(2,5):
    res_biKMeans.append(testBKM(dataArr,i,label_true))
```

```python
for index,k in enumerate(res_kMeans):
    print 'kM',index,':','%.06f'%(sum(k[0])/len(k[0])),'%.06f'%(sum(k[1])/len(k[1]))
for index,k in enumerate(res_biKMeans):
    print 'bKM',index,':','%.06f'%(sum(k[0])/len(k[0])),'%.06f'%(sum(k[1])/len(k[1]))
```

```python
for i in range(3):
    print stats.ttest_ind(res_kMeans[i][0],res_biKMeans[i][0])
for i in range(3):
    print stats.ttest_ind(res_kMeans[i][1],res_biKMeans[i][1])
```
